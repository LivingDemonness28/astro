# synastry

```python
from IPython.display import clear_output

def clear_console():
    clear_output(wait=True)

def synastry():
    planets = [
        "Sun", "Moon", "Mercury", "Venus", "Mars", "Jupiter", "Saturn", 
        "Uranus", "Neptune", "Pluto", "North Node", "South Node", 
        "Lilith", "Chiron", "Fortune", "Vertex", "Ascendant", "Descendant", "IC", "MC"
    ]
    signs = [
        "Aries", "Taurus", "Gemini", "Cancer", "Leo", "Virgo", 
        "Libra", "Scorpio", "Sagittarius", "Capricorn", "Aquarius", "Pisces"
    ]
    
    part1 = []
    degree1 = []
    minute1 = []
    part2 = []
    degree2 = []
    minute2 = []
    
    # Input for Partner 1
    print("Enter data for Partner 1:")
    for planet in planets:
        sign = input(f"Sign of {planet}: ").capitalize()
        
        while True:
            try:
                degree = int(input(f"Degree of {planet}: "))
                if 0 <= degree <= 29:
                    break
                else:
                    print("Invalid input. Degree must be between 0 and 29. Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")
        
        while True:
            try:
                minute = int(input(f"Minute of {planet} (must be between 0 and 59): "))
                if 0 <= minute <= 59:
                    break
                else:
                    print("Invalid input. Minute must be between 0 and 59. Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")

        part1.append(sign)
        degree1.append(degree)
        minute1.append(minute)
        print(" ")
    
    # Input for Partner 2
    print("Enter data for Partner 2:")
    for planet in planets:
        sign = input(f"Sign of {planet}: ").capitalize()
        
        while True:
            try:
                degree = int(input(f"Degree of {planet} (must be between 0 and 29): "))
                if 0 <= degree <= 29:
                    break
                else:
                    print("Invalid input. Degree must be between 0 and 29. Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")
        
        while True:
            try:
                minute = int(input(f"Minute of {planet} (must be between 0 and 59): "))
                if 0 <= minute <= 59:
                    break
                else:
                    print("Invalid input. Minute must be between 0 and 59. Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")

        part2.append(sign)
        degree2.append(degree)
        minute2.append(minute)
        print(" ")
    
    # Determine Ascendant and house mappings
    ascendant1 = part1[planets.index("Ascendant")]
    ascendant2 = part2[planets.index("Ascendant")]
    asc_index1 = signs.index(ascendant1)
    asc_index2 = signs.index(ascendant2)
    house_mapping1 = {signs[(asc_index1 + i) % 12]: i + 1 for i in range(12)}
    house_mapping2 = {signs[(asc_index2 + i) % 12]: i + 1 for i in range(12)}
    
    # Clear console before showing results
    clear_console()
    
    # **Store aspects in a list before printing**
    aspects_list = []

    # Check for aspects
    for i in range(len(planets)):
        for j in range(len(planets)):
            pos1 = signs.index(part1[i])
            pos2 = signs.index(part2[j])
            degree_diff = abs(degree1[i] - degree2[j])
            minute_diff = abs(minute1[i] - minute2[j])
            
            # Normalize to degrees and minutes
            total_diff_minutes = degree_diff * 60 + minute_diff
            orb_degree = total_diff_minutes // 60  # Whole degree
            orb_minute = total_diff_minutes % 60  # Remainder in minutes
            
            # Conjunction
            if part1[i] == part2[j] and total_diff_minutes <= 480:  # 8° in minutes
                aspects_list.append((f"{planets[i]} {part1[i]} {degree1[i]}° {minute1[i]}′ conjuncts {planets[j]} {part2[j]} {degree2[j]}° {minute2[j]}′, {orb_degree}° {orb_minute}′ orb", orb_degree, orb_minute))

            # Opposition
            elif (pos1 + 6) % 12 == pos2 and total_diff_minutes <= 480:  # 8° in minutes
                aspects_list.append((f"{planets[i]} {part1[i]} {degree1[i]}° {minute1[i]}′ opposes {planets[j]} {part2[j]} {degree2[j]}° {minute2[j]}′, {orb_degree}° {orb_minute}′ orb", orb_degree, orb_minute))

            # Square
            elif ((pos1 + 3) % 12 == pos2 or (pos1 - 3) % 12 == pos2) and total_diff_minutes <= 420:  # 7° in minutes
                aspects_list.append((f"{planets[i]} {part1[i]} {degree1[i]}° {minute1[i]}′ squares {planets[j]} {part2[j]} {degree2[j]}° {minute2[j]}′, {orb_degree}° {orb_minute}′ orb", orb_degree, orb_minute))

            # Sextile
            elif ((pos1 + 2) % 12 == pos2 or (pos1 - 2) % 12 == pos2) and total_diff_minutes <= 60:  # 1° in minutes
                aspects_list.append((f"{planets[i]} {part1[i]} {degree1[i]}° {minute1[i]}′ sextiles {planets[j]} {part2[j]} {degree2[j]}° {minute2[j]}′, {orb_degree}° {orb_minute}′ orb", orb_degree, orb_minute))

            # Trine
            elif ((pos1 + 4) % 12 == pos2 or (pos1 - 4) % 12 == pos2) and total_diff_minutes <= 480:  # 8° in minutes
                aspects_list.append((f"{planets[i]} {part1[i]} {degree1[i]}° {minute1[i]}′ trines {planets[j]} {part2[j]} {degree2[j]}° {minute2[j]}′, {orb_degree}° {orb_minute}′ orb", orb_degree, orb_minute))

    # **Sort aspects by orb value in ascending order (by minutes first, then degrees)**
    aspects_list.sort(key=lambda x: (x[1], x[2]))  # Sort by degree, then minute

    # **Print sorted aspects**
    print("Aspects in Synastry:")
    for aspect in aspects_list:
        print(aspect[0])  # Print only the formatted aspect string
```

```python
from IPython.display import clear_output

def clear_console():
    clear_output(wait=True)

def composite():
    planets = ["Sun", "Moon", "Mercury", "Venus", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune", "Pluto", "North Node", "South Node", "Lilith", "Chiron", "Ascendant", "Descendant", "IC", "MC"]
    signs = ["Aries", "Taurus", "Gemini", "Cancer", "Leo", "Virgo", 
             "Libra", "Scorpio", "Sagittarius", "Capricorn", "Aquarius", "Pisces"]
    
    # Define traditional and modern rulers
    rulers = {
        "Aries": {"traditional": "Mars", "modern": "Mars"},
        "Taurus": {"traditional": "Venus", "modern": "Venus"},
        "Gemini": {"traditional": "Mercury", "modern": "Mercury"},
        "Cancer": {"traditional": "Moon", "modern": "Moon"},
        "Leo": {"traditional": "Sun", "modern": "Sun"},
        "Virgo": {"traditional": "Mercury", "modern": "Mercury"},
        "Libra": {"traditional": "Venus", "modern": "Venus"},
        "Scorpio": {"traditional": "Mars", "modern": "Pluto"},
        "Sagittarius": {"traditional": "Jupiter", "modern": "Jupiter"},
        "Capricorn": {"traditional": "Saturn", "modern": "Saturn"},
        "Aquarius": {"traditional": "Saturn", "modern": "Uranus"},
        "Pisces": {"traditional": "Jupiter", "modern": "Neptune"}
    }
    
    placements = []
    degrees = []
    minutes = []
    aspects_list = []
    
    # Input for the composite chart
    print("Enter data for the composite chart")
    for planet in planets:
        sign = input(f"Sign of {planet}: ").capitalize()
        degree = int(input(f"Degree of {planet}: "))
        minute = int(input(f"Minute of {planet}: "))
        placements.append(sign)
        degrees.append(degree)
        minutes.append(minute)
        print(" ")
    
    clear_console()
    
    # Chart ruler
    ascendant_sign = placements[planets.index("Ascendant")]
    traditional_ruler = rulers[ascendant_sign]["traditional"]
    modern_ruler = rulers[ascendant_sign]["modern"]
    
    print("Composite Chart Ruler:")
    print(f"Ascendant: {ascendant_sign}")
    print(f"Traditional Ruler: {traditional_ruler}")
    print(f"Modern Ruler: {modern_ruler}")
    print("\n")
    
    # Check for aspects
    print("Aspects in the Composite Chart:")
    for i in range(len(planets)):
        for j in range(i + 1, len(planets)):  # Avoid duplicates by starting at `i + 1`
            pos1 = signs.index(placements[i])
            pos2 = signs.index(placements[j])
            degree_diff = abs(degrees[i] - degrees[j])
            minute_diff = abs(minutes[i] - minutes[j])
            
            total_diff_minutes = degree_diff * 60 + minute_diff
            orb_degree = total_diff_minutes // 60  # Whole degree
            orb_minute = total_diff_minutes % 60  # Remainder in minutes
            
            # Conjunction (same sign)
            if placements[i] == placements[j] and total_diff_minutes <= 480:
                aspects_list.append((f"{planets[i]} {placements[i]} {degrees[i]}° conjuncts {planets[j]} {placements[j]} {degrees[j]}° {orb_degree}°{orb_minute}' orb", orb_degree, orb_minute))

            # Opposition (180° apart)
            elif (pos1 + 6) % 12 == pos2 and total_diff_minutes <= 480:
                aspects_list.append((f"{planets[i]} {placements[i]} {degrees[i]}° oppose {planets[j]} {placements[j]} {degrees[j]}° {orb_degree}°{orb_minute}' orb", orb_degree, orb_minute))

            # Square (90° apart)
            elif ((pos1 + 3) % 12 == pos2 or (pos1 - 3) % 12 == pos2) and total_diff_minutes <= 420:
                aspects_list.append((f"{planets[i]} {placements[i]} {degrees[i]}° square {planets[j]} {placements[j]} {degrees[j]}° {orb_degree}°{orb_minute}' orb", orb_degree, orb_minute))

            # Sextile (60° apart)
            elif ((pos1 + 2) % 12 == pos2 or (pos1 - 2) % 12 == pos2) and total_diff_minutes <= 120:
                aspects_list.append((f"{planets[i]} {placements[i]} {degrees[i]}° sextile {planets[j]} {placements[j]} {degrees[j]}° {orb_degree}°{orb_minute}' orb", orb_degree, orb_minute))

            # Trine (120° apart)
            elif ((pos1 + 4) % 12 == pos2 or (pos1 - 4) % 12 == pos2) and total_diff_minutes <= 480:
                aspects_list.append((f"{planets[i]} {placements[i]} {degrees[i]}° trine {planets[j]} {placements[j]} {degrees[j]}° {orb_degree}°{orb_minute}' orb", orb_degree, orb_minute))
    # **Sort aspects by orb value in ascending order (by minutes first, then degrees)**
    aspects_list.sort(key=lambda x: (x[1], x[2]))  # Sort by degree, then minute

    # **Print sorted aspects**
    print("Aspects in Composite:")
    for aspect in aspects_list:
        print(aspect[0])  # Print only the formatted aspect string
```
